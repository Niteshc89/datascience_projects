```python
import requests
from bs4 import BeautifulSoup
from csv import writer
from time import sleep
from random import choice
import re
from requests.packages.urllib3.exceptions import InsecureRequestWarning
requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
import urllib.parse
import random
```

```python
# source for quotes 
base_url = "https://www.goodreads.com/quotes"
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/140.0.0.0 Safari/537.36"}
res = requests.get(f"{base_url}",headers=headers,verify=False)
pg_src = res.text
```

```python
# Scraping quotes using BeautifulSoup 
soup = BeautifulSoup(pg_src,"html.parser")
#soup
```

```python
all_quotes = []
quotes = soup.find_all(class_="quote")
#print(quotes[0])
for q in quotes[0:]:
    tag  = (q.find(class_="quoteText").text).strip()
    compo_label = re.split(r"\n",tag,re.MULTILINE)
    all_quotes.append({"quote":re.sub(r"[\“|\”]","",compo_label[0]),"author":compo_label[-1].strip()})
#len(all_quotes)
```

```python
quote = choice(all_quotes)
print("Here is the quote : ")
print(quote["quote"])
remaining_attempts = 4
```

```python
guess = ""

while quote["author"].lower() != guess.lower() and remaining_attempts > 0:
    guess = input(f"Who quoted this quote? Remaining attempts are {remaining_attempts}")
    remaining_attempts -=1 
    if guess.lower() == quote["author"].lower() :
        print("Congratulations ! You are correct.")
        break
    else :
        match remaining_attempts:
            case 3:
                print(f"Hint --> first letter of first name is {quote['author'][0]} and contains {len(quote['author'].split(' ')[0])} letters")
            case 2:
                print(f"Hint --> last letter of last name is {(quote['author'].split(' ')[-1])[0]} and contains {len(quote['author'].split(' ')[0])} letters")
            case 1:
                charlist = [i for i in quote['author'] if i!=' ']
                random.shuffle(charlist)
                print(f"Hint --> full name contains following combinations of letters :  {''.join(charlist)}")
            case _:
                print(f"You ran out of attempts. Correct answer is {quote['author']}")
```
